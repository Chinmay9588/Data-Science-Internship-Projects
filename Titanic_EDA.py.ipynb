```python
# =====================================
# ðŸ“Œ 1. Import Libraries
# =====================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# =====================================
# ðŸ“Œ 2. Load the Dataset
# =====================================
# Titanic dataset available directly in seaborn
df = sns.load_dataset('titanic')

# Preview first 5 rows
print("ðŸ”¹ First 5 Rows:")
display(df.head())

# Basic information about the dataset
print("\nðŸ”¹ Dataset Info:")
df.info()

# =====================================
# ðŸ“Œ 3. Initial Exploration
# =====================================
print("\nðŸ”¹ Shape of dataset:", df.shape)
print("\nðŸ”¹ Missing values per column:\n", df.isnull().sum())
print("\nðŸ”¹ Basic Statistics (Numeric Columns):")
display(df.describe())

# For categorical columns
print("\nðŸ”¹ Value counts for 'class':")
print(df['class'].value_counts())

# =====================================
# ðŸ“Œ 4. Data Cleaning
# =====================================
# Handle missing values
# Fill 'age' missing values with median
df['age'].fillna(df['age'].median(), inplace=True)

# Fill 'embarked' missing values with mode
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)

# Drop 'deck' column due to many missing values
df.drop(columns=['deck'], inplace=True)

# Verify changes
print("\nðŸ”¹ Missing values after cleaning:\n", df.isnull().sum())

# =====================================
# ðŸ“Œ 5. Univariate Analysis
# =====================================
# Age distribution
plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=30, kde=True, color='skyblue')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Gender count
sns.countplot(x='sex', data=df, palette='pastel')
plt.title('Passenger Count by Gender')
plt.show()

# Class distribution
sns.countplot(x='class', data=df, palette='Set2')
plt.title('Passenger Class Distribution')
plt.show()

# =====================================
# ðŸ“Œ 6. Bivariate Analysis
# =====================================
# Survival rate by gender
sns.countplot(x='sex', hue='survived', data=df, palette='muted')
plt.title('Survival Count by Gender')
plt.show()

# Fare distribution by class
sns.boxplot(x='class', y='fare', data=df, palette='coolwarm')
plt.title('Fare Distribution by Class')
plt.show()

# Age vs Fare colored by Survival
sns.scatterplot(x='age', y='fare', hue='survived', data=df, palette='Set1')
plt.title('Age vs Fare (Survival)')
plt.show()

# =====================================
# ðŸ“Œ 7. Correlation Analysis
# =====================================
# Convert 'survived' to int for correlation
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# =====================================
# ðŸ“Œ 8. Multivariate Analysis
# =====================================
sns.pairplot(df[['age', 'fare', 'survived']], hue='survived', palette='husl')
plt.suptitle('Pairplot of Age, Fare, and Survival', y=1.02)
plt.show()

# =====================================
# ðŸ“Œ 9. Key Insights
# =====================================
# Calculate survival rate by class and gender
survival_by_class_gender = df.groupby(['class','sex'])['survived'].mean()
print("\nðŸ”¹ Survival Rate by Class and Gender:\n", survival_by_class_gender)

# Average age and fare of survivors vs non-survivors
avg_values = df.groupby('survived')[['age','fare']].mean()
print("\nðŸ”¹ Average Age and Fare (Survived=1, Not=0):\n", avg_values)

# =====================================
# ðŸ“Œ 10. Save Cleaned Dataset (Optional)
# =====================================
df.to_csv('titanic_cleaned.csv', index=False)
print("\nâœ… Cleaned dataset saved as 'titanic_cleaned.csv'")
```
